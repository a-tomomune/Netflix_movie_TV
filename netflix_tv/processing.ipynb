# 前処理

## 日付変更

### date_added

```python
title["date_added"] = pd.to_datetime(title["date_added"], errors="coerce")
title["date_added"].head()
```

## 欠損値処理

```python
title["director"]=title["director"].fillna("Unknown")
title["director"].head()
```

```python
title["cast"]=title["cast"].fillna("Unknown")
title["country"]=title["country"].fillna("Unknown")
title["rating"]=title["rating"].fillna("Unknown")
title["duration"]=title["duration"].fillna("Unknown")
```

## カラム追加

### 追加年

```python
title["add_year"]=title["date_added"].dt.year
title["add_year"].head()
```

### 追加月

```python
title["add_month"]=title["date_added"].dt.month
title["add_month"]
```
